```python
# ミッドウィンター女性バラバラ殺人事件のpython解答
# 演習１：重回帰式を求める
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
# y：身長、x1：足の長さ、x2：腕の長さ、N：データ数、p:独立変数の自由度
y  = np.array([168, 152, 154, 173, 157, 159, 165, 161, 175, 163])
x1 = np.array([77, 65, 71, 79, 72, 73, 75, 73, 83, 69])
x2 = np.array([70, 58, 61, 69, 65, 61, 70, 69, 71, 65])
N = 10
p = 2
```

```python
# 統計量の計算
yy   = y * y
x12  = x1 * x1
x22  = x2 * x2
yx1  = y * x1
yx2  = y * x2
x1x2 = x1 * x2
```

```python
# 平方和と積和の計算。
syy = np.sum(yy) - np.sum(y) * np.sum(y) / N
sx1x1 = np.sum(x12) - np.sum(x1) * np.sum(x1) / N
sx2x2 = np.sum(x22) - np.sum(x2) * np.sum(x2) / N
syx1 = np.sum(yx1) - np.sum(y) * np.sum(x1) / N
syx2 = np.sum(yx2) - np.sum(y) * np.sum(x2) / N
sx1x2 = np.sum(x1x2) - np.sum(x1) * np.sum(x2) / N 
print(syy, sx1x1, sx2x2,syx1,syx2,sx1x2)
```

```python
# 平方和積和行列
ss_sp = np.array([[sx1x1,sx1x2],[sx1x2,sx2x2]])
print(ss_sp)
```

```python
# 逆行列
ss_sp_inv = np.linalg.inv(ss_sp)
print(ss_sp_inv)
```

```python
# 編相関係数
[b1, b2] = np.dot(ss_sp_inv, [syx1, syx2])
b0 = np.sum(y) / N - b1 * np.sum(x1) / N - b2 * np.sum(x2) / N
print(b1, b2, b0)
```

```python
# 描画
fig = plt.figure()
ax = mplot3d.Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)

ax.scatter3D(x1, x2, y)
linex = np.arange(60, 85)
liney = np.arange(60, 85)
z = b1 * linex + b2 * liney + b0
ax.set_xlabel("足の長さ")
ax.set_ylabel("腕の長さ")
ax.set_zlabel("身長")
ax.plot(linex, liney, z,"--")
plt.show()
```

```python
# 被害者の身長予測
z = b1 * 78 + b2 * 64 + b0
print(str(round(z,2)) + 'cm')
```

```python
# 演習２：重回帰式の検定をする
sr = b1 * syx1 + b2 * syx2
se = syy - sr
vr = sr / p
ve = se / (N - p - 1)
f0 = vr/ve
print(sr, se, vr, ve, N-p-1,f0)
# 有意水準0.05とすると、f0 ≒ 20.27 > F(2,7;0.05) ≒ 4.37となるので、仮設H0は棄却される
```

```python
# 演習３：決定係数の計算をする
rr = 1 - se / syy
rr_adj = 1 - ( N - 1 ) * se / (( N - p - 1 ) * syy)
r = np.sqrt(rr)
print(rr, rr_adj, r)
# 自由度調整済決定係数(rr_adj)及び重相関係数(r)が1に近いため、回帰式としては当てはまりがいいことがわかる
```

```python
# 演習4：偏回帰係数の検定をする。仮説H0をβ1=0、対立仮設H1をβ2≠0とする
t1 = b1 / np.sqrt(sx2x2 * ve / np.linalg.det(ss_sp) )
t2 = b2 / np.sqrt(sx1x1 * ve / np.linalg.det(ss_sp) )
print(t1, t2)
# t1 ≒ 92.47> t(7;0.05) ≒ 2.365となり、仮設H0は棄却される.ここから、足の長さx1は身長に影響を与えていることがわかる。
# また、t2 ≒ 1.45 < t(7;0.05) ≒ 2.365となり、仮設H0は棄却されない。よって腕の長さは身長に影響を与えているかは判断できない
```
